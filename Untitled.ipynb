```python
def support(price, q=0.50):
    
    df = pd.DataFrame(price.values, columns=['price']).reset_index()
    mod = smf.quantreg('price ~ index', df)
    res = mod.fit(q=.01)
    a0, a1 = res.params
    r2 = res.prsquared
    return(a0, a1, r2)
```

```python
df = pd.DataFrame(price.values, columns=['price']).reset_index()
mod = smf.quantreg('price ~ index', df)
res = mod.fit(q=.01)
```
